```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix,classification_report
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler, LabelEncoder
titanic_df=pd.read_csv(r"tested.csv")
titanic_df.head()
```

```python
titanic_df.describe()
```

```python
titanic_df["Age"]= titanic_df["Age"].fillna(titanic_df["Age"].median())
titanic_df.info()
```

```python
titanic_df["Fare"]= titanic_df["Fare"].fillna(titanic_df["Fare"].median())
titanic_df.info()
```

```python
titanic_df= titanic_df.drop("Cabin", axis=1)
```

```python
titanic_df.head()
```

```python
identity=titanic_df[["PassengerId","Name","Ticket"]]
identity.head()
```

```python
titanic_df=pd.get_dummies(titanic_df,columns=["Pclass","Embarked"],drop_first=True,dtype=int)
titanic_df.head()
```

```python
le= LabelEncoder()
titanic_df["Sex"]=le.fit_transform(titanic_df["Sex"])
titanic_df.head()
```

```python
gender_survived= titanic_df.groupby("Sex")["Survived"].agg("count")
gender_survived
```

```python
Y= titanic_df["Survived"]
X= titanic_df.drop(["Survived","PassengerId","Name","Ticket"], axis=1)
X.head()
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
lr=LogisticRegression()
model= lr.fit(X_train,Y_train)
```

```python
X.info()
```

```python
Y_pred= model.predict(X_test)
```

```python
#evaluating the model
accuracy_score(Y_pred,Y_test)
```

```python
precision_score(Y_pred,Y_test)
```

```python
recall_score(Y_pred,Y_test)
```
