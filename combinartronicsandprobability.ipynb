```python
def permutation (n,k):
  return math.factorial(n)/math.factorial(n-k)
```

```python
permutation (8,3)
```

```python
def combination(n,k):
  return math.factorial(n)/ (math.factorial(n-k)*math.factorial(k))
```

```python
combination (8,3)
```

```python
combination (10,5)
```
