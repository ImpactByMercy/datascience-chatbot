```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
# loading the data
sb_data = pd.read_csv(r"superbowl.csv")
sb_data.tail()
```

```python
sb_data["Coin Toss Winner"].value_counts()
```

```python
sb_data["Game Winner"].value_counts()
```

```python
coin_tost_counts=sb_data["Coin Toss Outcome"].value_counts()
```

```python
fair_coin_expected=[sum(coin_tost_counts)/2,sum(coin_tost_counts)/2]
fair_coin_expected
```

```python
fair_coin_observed=coin_tost_counts.values
fair_coin_observed
```

```python
#testing whether its a fair coin
fair_coin_result=stats.chisquare(fair_coin_observed,fair_coin_expected)
fair_coin_result
```

```python
gamewinner_counts=sb_data["Game Winner"].value_counts()
gamewinner_counts
```

```python
observed=gamewinner_counts.values
observed
```

```python
expected=[sum(gamewinner_counts)/2,sum(gamewinner_counts)/2]
expected
```

```python
results=stats.chisquare(observed,expected)
results
```

```python
nfl_games= pd.read_csv(r"nfl_games.csv")
nfl_games.head()
```

```python
nfl_games = nfl_games[(nfl_games["neutral"] == 0) & (nfl_games["playoff"] == 0) & (nfl_games["season"] >= 1966 )]
# Engineer a new feature
nfl_games.loc[nfl_games["result1"] == 0, "Game Winner"] = "Away Team"
nfl_games.loc[nfl_games["result1"] == 1, "Game Winner"] = "Home Team"

game_winner_counts_nfl = nfl_games["Game Winner"].value_counts().sort_index()
game_winner_counts_nfl
```

```python
nfl_observed= game_winner_counts_nfl.values
nfl_observed
```

```python
nfl_expected= [sum(game_winner_counts_nfl)/2,sum(game_winner_counts_nfl)/2]
nfl_expected
```

```python
nfl_chisquare= stats.chisquare(nfl_observed,nfl_expected)
nfl_chisquare
```

```python
independence_table = pd.crosstab(sb_data["Coin Toss Winner"], sb_data["Game Winner"])
independence_table
```

```python
chi2,p,dof,exf=stats.chi2_contingency(independence_table)
print(chi2,p,dof,exf)
```
