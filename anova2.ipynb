```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
it_salaries=pd.read_csv(r"IT_salaries.csv")
it_salaries.head()
```

```python
salaries= "S~+C(E)+C(M)+X"
lm=ols(salaries,it_salaries).fit()
table=sm.stats.anova_lm(lm,typ=2)
table
```

#### ONE WAY---TWO SAMPLE T TEST
# COMPARES ONE INDEPENDNT VARIABLE FACTOR with two or more groups eg comparing test scores between 3 teaching methods
#### TWO WAY ANOVA--
# compares two independent variables for example test scores by teaching method and gender
#### Repeated measures anova---
# compares same subject measured under different conditions or times eg blood pressure before, during, and after treatment on same patient 

```python
#one way
import pandas as pd
import numpy as np
from scipy.stats import f_oneway

# Sample data
campaign_A = [12, 15, 14, 10, 13, 15, 11, 14, 13, 16]
campaign_B = [18, 17, 16, 15, 20, 19, 18, 16, 17, 19]
campaign_C = [10, 9, 11, 10, 12, 9, 11, 8, 10, 9]
# Perform one-way ANOVA
f_stat,p_value=f_oneway(campaign_A,campaign_B,campaign_C)
print(f_stat,p_value)
```

```python
if p_value<0.05:
  print("We reject the null hypothesis")
else:
  print("We fail to reject the null hyothesis")
```
