```python
#mean of the samples
sample_size=2
combs=list(itertools.combinations(pumpkin_dict.keys(),sample_size))
combs
```

```python
def sample_mean(n,data):
  n=n #sample size
  mu=sum(data.values())/len(data)
  #genarate possible combinations using the given sample size 
  combinations=list(itertools.combinations(data.keys(),n))
  #sample mean
  xbar_list=[]
  #for loop
  for comb in combinations:
    sample_weights=[data[p]for p in comb]
    sample_mean=sum(sample_weights)/n
    xbar_list.append(sample_mean)
  return combinations,xbar_list


```

```python
combinations, xbar_list=sample_mean(sample_size,pumpkin_dict)
for c,m in zip(combinations,xbar_list):
  print(c,m)
```

```python
#z-test
from scipy import stats
import numpy as np
sample_size=40
population_mean=100
population_sd=16
x_bar=103
alpha=0.05
z_statistics=(x_bar-population_mean)/(population_sd/np.sqrt(sample_size))
z_statistics
```

```python
z_cdf=stats.norm.cdf(z_statistics)
z_cdf
```

```python
p_value=1-z_cdf
p_value
```

```python
if p_value>alpha:
  print("FAIL TO REJECT THE NULL HYPOTHESIS")
else:
  print("REJECT THE NULL HYPOTHESIS")
```

```python
from scipy import stats
import numpy as np
sample_size_secs=36
mean_sec=60
sec_sd=16
sec_x_bar=75
alternative_hypothesis="contradicting the claim of first sevice"
null_hypothesis="accepting the claim of first service"
z_sec_statistics=(sec_x_bar-mean_sec)/(sec_sd/np.sqrt(sample_size_secs))
z_sec_statistics
```

```python
z_sec_cdf=stats.norm.cdf(z_sec_statistics)
z_sec_cdf
```

```python
p_value_sec=1-z_sec_cdf
p_value_sec
```

```python
if p_value_sec>alpha:
  print(alternative_hypothesis)
else:
  print(null_hypothesis)
```
