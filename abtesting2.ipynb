```python
import pandas as pd 
import numpy as np
from scipy import stats
```

```python
website_data=pd.read_csv(r"homepage_actions.csv")
website_data.head()
```

```python
control_data=website_data[website_data["group"]=="control"]
control_data
```

```python
control_data["group"].value_counts()
```

```python
experimental_data=website_data[website_data["group"]=="experiment"]
experimental_data.tail()
```

```python
experimental_clicks=sum(experimental_data["action"]=="click")
experimental_clicks
```

```python
experimental_views=sum(experimental_data["action"]=="view")
experimental_views
```

```python
experimental_ctr=experimental_clicks/experimental_views
experimental_ctr
```

```python
control_clicks=sum(control_data["action"]=="click")
control_clicks
```

```python
control_views=sum(control_data["action"]=="view")
control_views
```

```python
control_ctr=control_clicks/control_views
control_ctr
```

```python
experimental_mean= len(experimental_data)*experimental_ctr
experimental_mean
```

```python
control_mean=len(control_data)*control_ctr
control_mean
```

```python
pooled_probability=(experimental_clicks+control_clicks)/(len(control_data)+len(experimental_data))
pooled_probability
```

```python
control_len=len(control_data)
control_len
```

```python
experimental_len=len(experimental_data)
experimental_len
```

```python
standard_error=np.sqrt((pooled_probability*(1-pooled_probability))*((1/experimental_len)+(1/control_len)))
standard_error
```

```python
z_statistic=(experimental_ctr-control_ctr)/standard_error
z_statistic
```

```python
pvalue=2*(1-stats.norm.cdf(abs(z_statistic)))
pvalue
```
