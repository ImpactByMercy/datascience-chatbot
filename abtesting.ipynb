```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
```

```python
homepage_data=pd.read_csv(r"homepage_actions.csv")
homepage_data.head()
```

```python
#filtering the experimental group
experimental_group = homepage_data[homepage_data["group"]=="experiment"]
experimental_group.head()
```

```python
click_action = experimental_group["action"].value_counts()
click_action
```

```python
control_group = homepage_data[homepage_data["group"]=="control"]
control_group.head()
```

```python
click_action_control= control_group["action"].value_counts()
click_action_control
```

```python
experimental_views= (experimental_group["action"]=="view").sum()
experimental_views
```

```python
experimental_clicks=(experimental_group["action"]=="click").sum()
experimental_clicks
```

```python
experimental_ctr= (experimental_clicks/experimental_views)
experimental_ctr
```

```python
control_views = (control_group["action"]=="view").sum()
control_views
```

```python
control_clicks= (control_group["action"]=="click").sum()
control_clicks
```

```python
control_ctr= (control_clicks/control_views)
control_ctr
```

```python
probability= (experimental_clicks+control_clicks)/(experimental_views+control_views)
probability
```

```python
e_sum= experimental_group["action"].count()
e_sum
```

```python
experimental_mean= (experimental_views*experimental_ctr)
experimental_mean
```

```python
control_mean =(control_views*control_ctr)
control_mean
```

```python
pooled_std= np.sqrt(probability*(1-probability)*((1/control_views)+(1/experimental_views)))
pooled_std
```

```python
z_statistic = (experimental_ctr-control_ctr)/(pooled_std)
z_statistic
```

```python
cdf= stats.norm.cdf(z_statistic)
cdf
```
