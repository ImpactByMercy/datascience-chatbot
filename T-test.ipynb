```python
import numpy as np
from scipy import stats #perform hyothesis testing

sample = [122.09, 100.64, 125.77, 120.32, 118.25,  
        96.47, 111.4 ,  80.66, 110.77, 111.14, 
        102.9, 114.54,  88.09,  98.59,  87.07, 
       110.43, 101.9 , 123.89,  97.03, 116.23, 
        108.3, 112.82, 119.57, 131.38, 128.39]
```

```python
sample_mean = np.mean(sample)
sample_mean
```

```python
population_mean =100
population_mean
```

```python
#based on what we have before was 100 and now from thw sample its 109 so it worked
#alternative = the training worked --- its the observetion we make first
#null = the training did not work --- its what we want to test
alternative_hypothesis = "THE TRAINING WORKED"
null_hypothesis = "THE TRAINING DID NOT WORK"
```

```python
aplha =0.05 #significance level
```

```python
sample_size = len(sample) #number of observations in the sample
sample_size
```

```python
sample_sd = np.std(sample, ddof=1) #sample standard deviation
sample_sd
```

```python
t_statistic=(sample_mean-population_mean)/(sample_sd/np.sqrt(sample_size))
t_statistic
```
