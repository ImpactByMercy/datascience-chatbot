```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
house_prices= pd.read_csv(r"house_prices_dataset.csv")
house_prices.head()
```

```python
house_prices.columns
```

```python
house_prices=house_prices.drop(columns=["zipcode","lat","long"],axis=1)
house_prices.head()
```

```python
formula="price~+C(grade)+C(view)+C(condition)+C(waterfront)+bedrooms+bathrooms+sqft_living+sqft_lot+floors+year_built+year_renovated"
lm=ols(formula,house_prices).fit()
table=sm.stats.anova_lm(lm,typ=2)
table
```
