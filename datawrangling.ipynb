#### Data Wrangling
involves filtering and sorting data frames, feature engineering, grouping $ aggregation, merging and joining Dataframes

```python
# filtering $ sorting
import pandas as pd
data=pd.read_csv(r"tested.csv")
data.head()
```

```python
data.info()
```

```python
data=data.drop("Cabin",axis=1)
```

```python
data.head()
```

```python
#filtering data --comparison operators ---- boolean indexing
# obtain the adults column
adults=data[data["Age"]>=18]
adults.head()
```

```python
data["Age"].mean()
```

```python
adults["Age"].mean()
```

```python
#only males
males=data[data["Sex"]=="male"]
males.head()
```

```python
males["Sex"].value_counts()
```

```python
#rows with Q(Embarked)
emba=data[data["Embarked"]=="Q"]
emba.head()
```

```python
## Sorting Data
#sort_values....is used to sort values
#by column Asc (true or False)
data=data.sort_values(by="Fare",ascending=True)
data.head()
```

```python
#descending 
data=data.sort_values(by="Fare",ascending=False)
data.head()
```

```python
##Feature Engineering creating new columns that helps in improving your analysing and models
data["Age_group"]=pd.cut(data["Age"],bins=[0,12,19,59,100],labels=["child","teen","adult","senior"])#cut(column,bins,labels)
data.head()
```

```python
#extracting titles from name 
data["Title"]=data["Name"].apply(lambda name: name.split(",")[1].split(".")[0].strip())
data.head()
```
